## Solution for exercise 1

```python
import timeit
import random
import matplotlib.pyplot as plt

yget = []
yup = []
reps = 10
rng = range(10, 10000, 10)
for i in rng:
    for j in range(reps):
        t = timeit.Timer("x.get(0)", "from __main__ import x")
        t2 = timeit.Timer("x.update({0:None})", "from __main__ import x")
        x = {j:None for j in range(i)}
        yget_time = t.timeit(number=10)
        yup_time = t2.timeit(number=10)
    yget.append(yget_time/reps)
    yup.append(yup_time/reps)
x = rng
plt.plot(x, yget, 'r', label="Get method")
plt.plot(x, yup, 'b', label="Update method")
plt.title("Performance of Get and Update Methods")
plt.xlabel("Size of Dictionary")
plt.ylabel("Time (in milliseconds)")
plt.ylim((0, 0.00002))
plt.legend()
plt.show()
```

**Discussion of the Results**

This shows that the performance of the get method and the update method are both O(1) in terms of performance. There are little errors in the graph, but overall the performance is consistent with regards to the size of the dictionary.


## Solution for Exercise 2

```python
import timeit
import random
import matplotlib.pyplot as plt

yl = []
yd = []
reps = 10
rng = range(100, 100000, 100)
for i in rng:
    for j in range(reps):
        t = timeit.Timer("del x[0]", "from __main__ import x")
        x = list(range(i))
        yl_time = t.timeit(number=1)
        x = {j:None for j in range(i)}
        yd_time = t.timeit(number=1)
    yl.append(yl_time/reps)
    yd.append(yd_time/reps)
x = rng
plt.plot(x, yd, 'r', label="Dictionary")
plt.plot(x, yl, 'b', label="List")
plt.title("Performance Comparison of Del Operator (List and Dictionary)")
plt.xlabel("Size")
plt.ylabel("Time")
plt.legend()
plt.show()
```

**Discussion of the Results**

This graph shows that the performance of the del operator on lists is O(n) and the performance on dictionaries is O(1). Overall, the errors in the graph are negligible as the shape of the graph shows.
